# SVM

In diesem Notebook wird erklärt wie man eine `SVM` mit Python umsetzt.  
Das ganze wird anhand eines einfachen Datensatzes durchexerziert. In diesem Datensatz gibt es drei Variablen. `Age` und `Interest` werden benutzt um eine Vorhersage darüber zu treffen, ob ein gewisses Produkt gekauft wurde oder nicht. (`success` mit Wert 1 oder 0)

```python
import pandas as pd

df = pd.read_csv("./classification.csv")
df.count()
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X = df[["age", "interest"]].values
y = df["success"].values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0, test_size = 0.25)
```

Die Daten müssen bei `SVM` immer skaliert werden. Das hängt grundsätzlich mit dem `Kernel` zusammen.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

Es gilt hierbei:
<div style="font-size: 200%; margin-top: 10px;">$ \gamma ={\tfrac {1}{2\sigma ^{2}}} $</div>

$\sigma$ kann innerhalb des Aufrufs des `svm` übergeben werden. Die obere Funktion beschreibt nur die Umrechnung von $\sigma$ in $\gamma$

Hier beginnt die eigentliche `SVM` Funktion. Wichtig dabei zu wissen ist, dass die Parameter `kernel`, `gamma` und `C` variiert werden können. Später empfiehlt es sich diese mit einem `gridsearch` zu optimieren.

```python
from sklearn.svm import SVC
import numpy as np

model = SVC(kernel = "rbf", gamma = 1, C = 1000, probability=True)
model.fit(X_train, y_train)

print(model.score(X_test, y_test))
```

**Veranschaulichung und Beschreibung der Parameter $\gamma$ und `C`**:

![Gamma und C](./gammaC.png)

**Links-Oben:**
> 'Hügel' sind eher flach da $\gamma$ klein ist. Das führt dazu dass sich der allgemeine Trend ausbreiten kann.

**Rechts-Oben:**
> 'Hügel' sind sehr steil da $\gamma$ groß ist. Das führt dazu dass der allgemeine Trend verloren geht.

**Oben-Mitte:**
> C ist klein, das führt dazu, dass der allgemeine Trend wichtig ist. Dort wo viele blaue / rote Punkte sind bildet sich ein Trend. Einzelne Werte fallen nicht wirklich ins Gewicht. Man kann sagen bei einem kleinen C möchte man sich ganz sicher sein, dass man richtig zuteilt. Man braucht mehr als einen oder zwei Punkte um einen Trend abzubilden.

**Unten Mitte:**
> C ist groß, das führt dazu, dass der allgemeine Trend nicht wichtig ist. Einzelne Werte fallen ins Gewicht. Überall wo auch nur ein einziger blauer bzw. roter Punkt ist, bildet sich ein Trend. Orientierung an einzelnen Punkten!

----

Parameter C:
>Parameter bei der Erstellung der Trennebene

*Hohes C:*
* Einzelne Punkte sind wichtig, allgemeiner Trend ist weniger wichtig (Overfitting)

*Niedriges C:*
* Einzelne Punkte sind weniger wichtig, allgemeiner Trend ist wichtiger (Underfitting)

-------

Parameter $\gamma$:
>Steht uns nur beim `RBF-Kernel` zur Verfügung, Parameter zur Erstellung der 'Hügel'

*Hohes Gamma:*
* Einzelne Punkte sind wichtig, allgemeiner Trend ust weniger wichtig (Overfitting)

*Niedriges Gamma:*
* Einzelne Punkte sind weniger wichtig, allgemeiner Trend ist wichtiger (Underfitting)

## Hyperparameter optimieren

Zuerst wird die Methode der `pipeline` vorgestellt:

* Innerhalb der Pipeline kann man mehrere Tupel abspeichern. Dabei kann man den Namen (in Anführungszeichen) frei wählen.


```python
from sklearn.pipeline import Pipeline
```

```python
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", SVC(kernel = "rbf", gamma = 0.01, C = 1, probability=True))
])
```

```python
pipeline.fit(X, y)
```

Diese gerade erstellte Pipeline kann nun als ganz normales Modell verwendet werden. Der Vorteil ist hier nun, dass man anstatt von `X` und `y`(wie oben verwendet) auch die gesplitteten Daten übergeben kann:

```python
pipeline.fit(X_train, y_train)
```

```python
pipeline.score(X_test, y_test)
```

## GridSearch()

Mit GridSearch kann man die in der Piepline definierten Variablen nachträglich verändern und optimieren.

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "svm__kernel":['rbf', 'poly', 'sigmoid'],
    "svm__gamma":[0.001, 0.01, 0.1, 1, 10],
    "svm__C":[0.001, 0.01, 1, 10]
}
```

```python
clf = GridSearchCV(pipeline, param_grid)
```

```python
clf.fit(X_train, y_train)
```

```python
print(clf.best_params_)
```

```python
print(clf.score(X_test, y_test))
```
