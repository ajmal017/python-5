```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier, AdaBoostClassifier
from  sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
import pandas as pd
import matplotlib.pyplot as plt
#import rpy2.robjects as ro
#from rpy2.robjects.packages import importr
#from rpy2.robjects import pandas2ri
#from rpy2.robjects.conversion import localconverter
from IPython.display import display

import warnings
warnings.simplefilter('ignore')

#pandas2ri.activate()
```

# Aufgabe 1

```python
data = pd.read_csv("spambase.csv", sep=",").dropna()
print(len(data))
display(data.head(10))
```

```python
x = data.iloc[:,:-1].values
y = data['Target'].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.8, random_state=0)
```

```python
def grid_search(model, x_train, y_train, x_test, y_test, params):
    clf = GridSearchCV(model, params, cv=10)
    clf.fit(x_train, y_train)
   
    print("Best parameters set found on development set:")
    print()
    print(clf.best_params_)
    print()
    print("Grid scores on development set:")
    print()
    means = clf.cv_results_['mean_test_score']
    stds = clf.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, clf.cv_results_['params']):
        print("%0.3f (+/-%0.03f) for %r"
              % (mean, std * 2, params))
    print()
    print("Detailed classification report:")
    print()
    print("The model is trained on the full development set.")
    print("The scores are computed on the full evaluation set.")
    print()
    y_true, y_pred = y_test, clf.predict(x_test)
    print(classification_report(y_true, y_pred))
    print()
    return pd.DataFrame(clf.cv_results_)
```

```python
tuned_params = [{"n_estimators": range(100,200, 10)}]
grid_search(BaggingClassifier(DecisionTreeClassifier()), x_train, y_train, x_test, y_test, tuned_params)
grid_search(RandomForestClassifier(), x_train, y_train, x_test, y_test, tuned_params)
grid_search(AdaBoostClassifier(DecisionTreeClassifier()), x_train, y_train, x_test, y_test, tuned_params)
```
