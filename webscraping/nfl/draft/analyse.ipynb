# Combine Seasons

```python
offense_2018.columns
```

```python
result = {}

for i in range(0, len(offense_2019)):
    t_2019 = offense_2019.iloc[i]['Tm']
    v_2019 = offense_2019.iloc[i][['G', 'PF', 'Yds', 'Ply', 'Y/P', 'TO', 'FL', '1stD', 'Cmp_p', 'Att_p',
       'Yds_p', 'TD_p', 'Int_p', 'NY/A_p', '1stD_p', 'Att_r', 'Yds_r', 'TD_r', 'Y/A_r', '1stD_r',
       'penalties', 'Yds_penalties', '1st_penalties', 'Sc%', 'TO%', 'EXP']]
    
    v_2018 = offense_2018[offense_2018['Tm'] == t_2019][['G', 'PF', 'Yds', 'Ply', 'Y/P', 'TO', 'FL', '1stD', 'Cmp_p', 'Att_p',
       'Yds_p', 'TD_p', 'Int_p', 'NY/A_p', '1stD_p', 'Att_r', 'Yds_r', 'TD_r', 'Y/A_r', '1stD_r',
       'penalties', 'Yds_penalties', '1st_penalties', 'Sc%', 'TO%', 'EXP']]
    
    resulting_df = (v_2019 + v_2018) / 2
    result[t_2019] = resulting_df
    
```

```python
offense = pd.DataFrame()
for key, val in result.items():
    
    val['team'] = key
    offense = offense.append(val)
```

```python
offense.set_index('team', inplace=True)
```
