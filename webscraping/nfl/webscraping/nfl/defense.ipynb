```python
import seaborn as sns
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

# Total Defense Crawl

```python
url ='https://www.espn.com/nfl/stats/team/_/view/defense/stat/total'

headers= {'User-Agent': 'Mozilla/5.0'}

response = requests.get(url)
print(response.status_code)

soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# stat_table =soup.find_all('table', class_ = "Table2__table-scroller Table2__right-aligned Table2__table")
stat_table =soup.find_all('table', class_ = "Table Table--align-right")
len(stat_table)

#name_table = soup.find_all('table', class_="Table2__right-aligned Table2__table-fixed Table2__Table--fixed--left Table2__table")
name_table = soup.find_all('table', class_="Table Table--align-right Table--fixed Table--fixed-left")
len(name_table)

stat_table = stat_table[0]
name_table = name_table[0]
```

```python
team_names = []
stat_headers = []
team_stats = []
temp = []

for row in name_table.find_all('tr'):
    for cell in row.find_all('td'):
        #print(cell.text)
        team_names.append(cell.text)

for row in stat_table.find_all('tr'):
    #print(cell.text)
    for cell in row.find_all('th'):
            #print(cell.text)
            stat_headers.append(cell.text)        

for row in stat_table.find_all('tr'):
    for cell in row.find_all('td'):
        temp.append(cell.text)
    team_stats.append(temp)
    temp = []
            
# print team names
team_names

# Need to remove some columns
stat_headers = stat_headers[5:]
stat_headers

# Team stats - First two empty lists items need to be removed
team_stats = team_stats[2:]
team_stats
```

```python
# Create a DataFrame

total_defense_stats = pd.DataFrame(team_stats, columns=stat_headers)
total_defense_stats['Team'] = team_names
total_defense_stats.set_index('Team', inplace = True)
```

```python
total_defense_stats
```
