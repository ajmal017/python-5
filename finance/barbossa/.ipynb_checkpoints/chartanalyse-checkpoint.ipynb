## Moving Average

```python
df['MA'] = df['Close'].rolling(window=20).mean()
```

```python
df['STD'] = df['Close'].rolling(window=20).std(ddof=0)
```

```python
df.tail()
```

```python
df['upper'] = (df['MA'] + 2 * df['STD'])
```

```python
df['lower'] = (df['MA'] - 2 * df['STD'])
```

```python
fig.add_trace(go.Scatter(x=df.index, y=df['MA'], name='Moving Average',
                         line = dict(color='royalblue', width=1, dash='dash')));
fig.add_trace(go.Scatter(x=df.index, y=df['lower'], name='Lower Limit',
                         line = dict(color='red', width=1, dash='dash')));
fig.add_trace(go.Scatter(x=df.index, y=df['upper'], name='Upper Limit',
                         line = dict(color='green', width=1, dash='dash')));

fig.show();
```

```python
def get_fibonacci(fib):
    idx_min = fib['Close'].idxmin()
    min_date = fib.iloc[idx_min]['Date']
    min_val = fib.iloc[idx_min]['Close']
    
    fib_high = fib[fib['Date'] > min_date]
    
    idx_max = fib_high['Close'].idxmax()
    max_date = fib.iloc[idx_max]['Date']
    
    val_618 = (min_val * 0.618 + min_val)
    val_50 = (min_val * 0.50 + min_val)
    val_382 = (min_val * 0.382 + min_val)
    
    print('Low: {} and High: {}'.format(min_date, max_date))
    print(str(val_618) + ' ' + str(val_382) + ' ' + str(val_50))
    return val_382, val_50, val_618
```

```python
val_382, val_50, val_618 = get_fibonacci(fib)
```

```python
fib = fib.tail(60)

fig = go.Figure(data=[go.Candlestick(x=fib.index,
                open=fib['new_open'],
                high=fib['High'],
                low=fib['Low'],
                close=fib['Close'])])
fig.show();
```

```python
fig.add_shape(
        # Line Horizontal
            type="line",
            x0=fib.index.min(),
            y0= val_382,
            x1=fib.index.max(),
            y1=val_382,
            line=dict(
                color="LightSeaGreen",
                width=1,
                dash="dashdot",
            ),
)
fig.add_shape(
        # Line Horizontal
            type="line",
            x0=fib.index.min(),
            y0= val_50,
            x1=fib.index.max(),
            y1=val_50,
            line=dict(
                color="LightSeaGreen",
                width=1,
                dash="dashdot",
            ),
)
fig.add_shape(
        # Line Horizontal
            type="line",
            x0=fib.index.min(),
            y0= val_618,
            x1=fib.index.max(),
            y1=val_618,
            line=dict(
                color="LightSeaGreen",
                width=1,
                dash="dashdot",
            ),
)
fig.show();
```
