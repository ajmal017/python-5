# ML Algorhytmus Test

Die `pandas` Library bietet mit der Methode `loc` schon eine vorgefertigte Funktion um kleinere ML-Modelle zu erstellen. Hier ein erstes Beispiel:  
Da wir die Daten gesehen haben, sagen wir einfach vorher, dass wenn die untersuchte Person eine Frau war, sie auch überlebt hat.

```python
train = pd.read_csv("train.csv")
```

```python
# Neue Spalte zum Datensatz hinzufügen und komplett mit 0 füllen
train["Hypothese"] = 0

#Nun wird der komplette Datensatz durchgegangen und alle Damen die Hypothesen Spalte auf 1 gesetzt (überlebt)
train.loc[train.Sex == "female", "Hypothese"] = 1

train["Resultat"] = 0
train.loc[train.Survived == train["Hypothese"], "Resultat"] = 1

print(train["Resultat"].value_counts(normalize=True))
```

# Daten säubern

```python
# In der Age Salte fehlten Werte
# Mit der unteren Funktion werden leere Einträge mit dem Mittelwert ersetzt

df["Age"]=df["Age"].fillna(df["Age"].dropna().median())
print(df.count())
```

```python
# male und femal in 1 und 0 umwandeln

df.loc[df["Sex"]=="male", "Sex"] = 0
df.loc[df["Sex"]=="female", "Sex"] = 1
```

```python
df["Embarked"] = df["Embarked"].fillna("S")
print(df.count())
```

```python
df.loc[df["Embarked"]=="S", "Embarked"] = 0
df.loc[df["Embarked"]=="C", "Embarked"] = 1
df.loc[df["Embarked"]=="Q", "Embarked"] = 2
```

# ML in Action

```python
# %load data_cleaning.py
def data_clean(data):
    data["Age"]=data["Age"].fillna(data["Age"].dropna().median())

    # male und femal in 1 und 0 umwandeln

    data.loc[data["Sex"]=="male", "Sex"] = 0
    data.loc[data["Sex"]=="female", "Sex"] = 1

    data["Embarked"] = data["Embarked"].fillna("S")
    data.loc[data["Embarked"]=="S", "Embarked"] = 0
    data.loc[data["Embarked"]=="C", "Embarked"] = 1
    data.loc[data["Embarked"]=="Q", "Embarked"] = 2
```

```python
from data_cleaning import data_clean

train = pd.read_csv("train.csv")
data_clean(train)

target = train[["Survived"]].values
features = train[["Pclass", "Age", "Sex", "SibSp", "Parch"]].values

classifier = linear_model.LogisticRegression()
model = classifier.fit(features, target)

print(model.score(features, target))
    
```
